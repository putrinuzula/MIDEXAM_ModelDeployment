Dropping the columns that we don't need

## EDA & Pre Processing

Mengecek distribusi data untuk setiap variable

```python
for column in df.columns:
    plt.figure(figsize = (8,5))
    sns.displot(df[column])
    plt.show()
```

### Spliting Data

```python
input = df.drop(['churn'], axis = 1)
output = df['churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input, output, test_size = 0.2, random_state = 42)
```

#### Data Train Check

```python
x_train.info()
```

- Dari hasil output di atas menunjukan bahwa 'CreditScore' hanya memiliki 33000, sedangkan variable lainnya memiliki 33007 data. Hal tersebut menunjukan bahwa 'CreditScore' memiliki missing value dan akan saya lakukan imputasi menggunakan mean.
- Variabel 'Gender' dan 'Geography' memiliki tipe data categorical, maka dari itu haru dilakukan encoding.

```python
x_train['CreditScore'] = x_train['CreditScore'].fillna(x_train['CreditScore'].mean())
```

```python
x_train.isnull().sum()
```

Melakukan label encoding untuk variable non-numeric

```python
from sklearn.preprocessing import LabelEncoder

encode = LabelEncoder()

x_train['Geography'] = encode.fit_transform(x_train['Geography'])
x_train['Gender'] = encode.fit_transform(x_train['Gender'])
```

```python
x_train.dtypes
```

```python
# Check Outlier
for column in x_train:
      plt.figure()
      x_train.boxplot([column])
```

Terdapat beberapa variable yang memiliki outliers, tetapi masih masuk akal seperti credit score serta age karena tetap pada variasinya.

```python
y_train.isnull().sum()
```

#### Data Testing

Melakukan pre-processing yang sama pada data testing

```python
x_test.info()
```

Melakukan imputasi missing value pada 'CreditScore' dengan menggunakan nilai mean dari x_train

```python
x_test['CreditScore'] = x_test['CreditScore'].fillna(x_train['CreditScore'].mean())
```

```python
x_test.isnull().sum()
```

```python
x_test.duplicated().sum()
```

```python
#Label encoding

x_test['Geography'] = encode.fit_transform(x_test['Geography'])
x_test['Gender'] = encode.fit_transform(x_test['Gender'])
```

```python
x_test.dtypes
```

```python
# Check Outlier
for column in x_test:
      plt.figure()
      x_test.boxplot([column])
```

## Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

RF_class = RandomForestClassifier(criterion= 'gini',n_estimators = 100,max_depth = 4)
RF_class.fit(x_train, y_train)
```

```python
y_predictRF = RF_class.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print('\nClassification Report with RF\n')
print(classification_report(y_test, y_predictRF, target_names=['1','2']))
```

## XG Boost Model

```python
from xgboost import XGBClassifier

xgb_class = XGBClassifier(max_depth = 4, n_estimators = 100)
xgb_class.fit(x_train, y_train)
```

```python
y_predictXGB = xgb_class.predict(x_test)
```

```python
y_predictXGB = xgb_class.predict(x_test)

print('\nClassification Report with XGB\n')
print(classification_report(y_test, y_predictXGB, target_names=['1','2']))
```

```python
import pickle as pkl
filename = 'XGB_class.pkl'
pkl.dump(xgb_class, open(filename, 'wb'))
```

Dapat disimpulkan bahwa Klasifikasi data 'Churn' menggunakan model XGBoost menghasilkan performa yang lebih baik dengan tingkat akurasi menyentuh 86%, dibandingkan dengan model Random Forest yang memiliki akurasi pada 84%. Kedua model tersebut dapat memprediksi secara baik diantara kelas yang ada dengan mencapai sekitar 95% ke atas.
